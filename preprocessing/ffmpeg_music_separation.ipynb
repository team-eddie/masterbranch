```python
import subprocess
import os
```

```python
path_dir = "G:\\팀 드라이브\\recommendation_project\\origin_video"
file_list=os.listdir(path_dir)
file_list.sort()
print(file_list)
```

```python
title=[]
for i in file_list:
    i_split=i.split(".")
    title.append(i_split[0])
print(title)
```

```python
for i in title:
    cmd = 'ffmpeg -y -i "G:\\팀 드라이브\\recommendation_project\\origin_video\\{}.mp4" -vn -acodec libmp3lame -ar 44.1k -ac 2 -ab 192k -f wav "G:\\팀 드라이브\\recommendation_project\\video_origin_sound\\audio1\\{}.wav"'.format(i,i)
    subprocess.call(cmd, shell=True)
    print(cmd)
```
